#### 01. 쓰기모드

````python
# fp = open("c:\\my\\text.txt", 'w')
# fp = open(r"c:\my\text.txt", 'w') # r'' - 제거문자( \t 등등 ) 무시
fp = open(r"test.txt", 'w') # 경로 없으면 현제 경로 / text.txt 파일이 있으면 새로 만든다 -> 덮어쓰기
print(type(fp))

fp.write('hello\n')
fp.write('korea')
fp.close()
````

````python
with open(r'test.txt','w') as fp:
    fp.write('hello2\n')
    fp.write('korea2')
    print(fp.closed)

print(fp.closed)
````

````python
with open(r'test.txt','w') as fp:
    fp.write('abcdefghi')
````

````python
fp = open(r'test.txt', 'r')
txt = fp.read()
print(txt)
fp.close()
````

````python
with open(r'test.txt','w') as fp:
    fp.write('abc\n')
    fp.write('def\n')
    fp.write('ghi')
````

````python
with open('test.txt', 'r') as fp:
    # txt = fp.read()
    # print(txt)
    txt = fp.readline()
    print(txt.strip()) # 'abc\n' --> 'abc'
    txt = fp.readline()
    print(txt)
    txt = fp.readline()
    print(txt)
````

````python
with open('test.txt', 'r') as fp:
    for r in fp:
        print(r.strip())
````

#### 02. csv 파일 기록

````python
import csv

fp = open('my.csv', 'w', newline='')
wr = csv.writer(fp)
# print(type(wr))
wr.writerow(['aa',10,20])
wr.writerow(['bb',10,20])
wr.writerow(['cc',10,20])
fp.close()
````

````python
fp = open('my.csv')
rd = csv.reader(fp)
for a,b,c in rd:
    print(a,b,c)
fp.close()
````
