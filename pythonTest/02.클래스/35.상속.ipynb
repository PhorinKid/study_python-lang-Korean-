```python
class People:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def setName(self, name):
        self.name = name

class Student(People):
    def __init__(self, name, age, stdNum):
        super().__init__(name, age) # super().__init__(self, name, age)
        self.stdNum = stdNum

    def show(self):
        print('이름 :', self.name)
        print('나이 :', self.age)
        print('학번 :', self.stdNum)

class Manager(People):
    def __init__(self, name, age, dept):
        super().__init__(name, age) # super().__init__(self, name, age)
        self.stdNum = dept

std = Student('홍길동', 20, '20250301')
std.setName('이순신')
std.show()
```
