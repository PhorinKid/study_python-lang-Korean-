#### 상위클래스에 정의된 함수를 하위클래스에서 재정의 하는 것

```python
class pet:
    def Sleep(self):
        print('zzz')
    def Eat(self):
        print('???')

class Dog(pet):
    def Speak(self):
        print("bow wow")
    def Eat(self):
        print('bone')

class Cat(pet):
    def Speak(self):
        print('meow')
    def Eat(self):
        print('fish')

dog = Dog()
dog.Eat()
dog.Speak()
dog.Sleep()

cat = Cat()
cat.Eat()
cat.Speak()
cat.Sleep()
```

```python
class MyFilter:
    def filterfn(self,v):
        pass
    def myfilter(self,data):
        return [n for n in data if self.filterfn(n)]

class MyFilter30t(MyFilter):
    def filterfn(self,v):
        return v>=30
    
class MyFilter20t(MyFilter):
    def filterfn(self,v):
        return v<=20
    
mylist = [10,20,30,40,50]
obj = MyFilter30t()
print(obj.myfilter(mylist))

obj1 = MyFilter20t()
print(obj1.myfilter(mylist))
```
