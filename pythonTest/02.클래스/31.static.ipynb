````python
class Test:
    st = 100 # static 멤버 (객체 생성X), 클래스 변수
    def __init__(self):
        self.a = 0 # instance 변수(반드시 객체 생성 필요)
        self.b = 0

    def setData(self, x, y): #instance 
        self.a = x
        self.b = y
    
    def show(self):
        print(self.a, self.b)

    @staticmethod # 데코레이터 (함수에 추가기능)
    def stFn(aa): # self가 없는 함수 (static 멤버함수), instance 변수 사용불가
        # self.a = 10 -> 불가능
        Test.st = aa

    @staticmethod
    def getSt():
        return Test.st

print(Test.st)
obj = Test()
obj.setData(10,20)
obj.show()

Test.stFn(30)
print(Test.st)
print(Test.getSt())
````

- 클래스 내부에 선언되어있고, 멤버함수 외부에 선언된 변수 - static변수
- self 가 없는 멤버함수는 static 멤버함수
- 클래스명.멤버함수()로 호출한다.

- 인스턴스 멤버데이터는 객체 생성시 메모리 별도로 할당되기 때문에 일반적으로 데이터 용도로 사용
- ststic 멤버는 연산 및 처리용도로 사용

````python
class Student:
    def __init__(self, name,age,addr):
        self.name = name
        self.age = age
        self.addr = addr

    def show(self):
        print('이름 :', self.name)
        print('나이 :', self.age)
        print('주소 :', self.addr)

std1 = Student('홍길동', 20, '서울시')
std1.show()
std2 = Student('임꺽정', 30, '부산시')
std2.show()
````

````python
mylist = []
mylist.append(Student('홍길동', 20, '서울시'))
mylist.append(Student('임꺽정', 30, '부산시'))

for n in mylist:
    n.show()
    print('------------')
````

````python
class Circle:
    def circle_area(self, r):
        return 3.14*r**2
    
    def cylinder(self, r,h):
        return 3.14*r**2*h

c = Circle()
print(c.circle_area(5))
print(c.cylinder(3,10))
````

````python
class Circle: # 동일한 기능의 함수를 묶음(생성자 불필요) - staticmethod / 단순 계산 등등 데이터 불필요
    @staticmethod
    def circle_area(r):
        return 3.14*r**2
    
    @staticmethod
    def cylinder(r,h):
        return 3.14*r**2*h

print(Circle.circle_area(5))
print(Circle.cylinder(3,10))
````
