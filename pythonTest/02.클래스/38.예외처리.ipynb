````python
pizz_slace = 8
person = 2
if person==0:
    print('사람수는 0이면 안됩니다.')
elif person<0:
    print('사람수는 양수여야 합니다.')
else:
    per_pizz_slice = pizz_slace/person
    print(per_pizz_slice)
````

#### if 문이 많아지면 유지보수가 어렵다.

````python
pizz_slace = 8
person = -2
try:
    if person<0:
        raise Exception('person 이 음수면 안됩니다.')
    per_pizz_slice = pizz_slace/person
    print(per_pizz_slice)
except ZeroDivisionError as err:
    print('예외발생', err)
except Exception as err: # 모든 예외를 모두 처리
    print('에러발생', err)
````

````python
n = input('숫자입력')
try:
    rst = int(n) + 10
    print(rst)
except ValueError as err:
    print('에러 발생 (반드시 숫자로 입력)', err)
````
