```python
class Test: # class Test(object):
    def __init__(self,a,b):
        self.a = a
        self.b = b

    def setData(self,x,y):
        self.a = x
        self.b = y
    def show(self):
        print(self.a, self.b)
    def __repr__(self):
        return f'a={self.a} b={self.b}'

    def __add__(self,other):
        print('add call')
        return self.a + other
    
    def __sub__(self,other):
        print('minus call')
        return self.b - other
```

```python
# mylist = list([10,20])
mylist = [10,20]
print(mylist)
```

```python
obj = Test(10,20)
obj.show()
print(obj) # obj.__repr__()

rst = obj + 11 # obj.__add__(11)
print(rst)

rst = obj - 11 # obj.__sub__(11)
print(rst)
```

```python
class MyDic:
    def __init__(self):
        self.mylist = {}

    def __setitem__(self,key,value):
        print("setitem call")
        self.mylist[key] = value

    def __getitem__(self, key):
        return self.mylist[key]


obj = MyDic()
obj['aa'] = 10 # obj.__setitem__('aa', 10)
print(obj['aa'])
```

#### iterator

```python
class MyIter:
    def __init__(self):
        self.nCnt = -1
        self.data = [11,22,33,44,55]

    def __iter__(self):
        print('iter call')
        return self # 리턴값은 self 고정
    
    def __next__(self):
        self.nCnt += 1
        if self.nCnt==5:
            raise StopIteration() # 반복문 종료
        return self.data[self.nCnt]
    
it = MyIter()
rst = next(it) # it.__next__()
print(rst)
rst = next(it)
print(rst)
rst = next(it)
print(rst)
rst = next(it)
print(rst)
rst = next(it)
print(rst)
```

```python
it = MyIter()
# it.__iter__() # 초기화 작업
# n = next(it)
# n = next(it)
# ...

for n in it:
    print(n)
```
