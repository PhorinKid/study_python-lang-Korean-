```python
class Test :
    def __init__(self,a,b):
        self.a = a
        self.b = b

    def setData(self,x,y):
        self.a = x
        self.b = y
    def show(self):
        print(self.a, self.b)

obj = Test(10,20)
obj.show()
obj.setData(100,200)
obj.show()
```

```python
class Test :
    def __init__(self,a=0,b=0):
        self.a = a
        self.b = b

    def setData(self,x,y):
        self.a = x
        self.b = y
    def show(self):
        print(self.a, self.b)

obj = Test(b=200)
obj.show()
obj.setData(100,200)
obj.show()
```
